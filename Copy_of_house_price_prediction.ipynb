##House Price Prediction

```python
 # Generic Libraries
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Data Set
from sklearn.datasets import fetch_california_housing
```

```python
# Get the data set, so we can get the feaure functions as well
data = fetch_california_housing()
```

```python
print(data.DESCR)
```

```python
data.data.shape
```

```python
# Created a dataframe and Added this data to the dataframe
# Independent Data
df = pd.DataFrame(data = data.data, columns = data.feature_names)
df.head()
```

```python
# Dependent Data
df['Target'] = data.target
df.head()
```
